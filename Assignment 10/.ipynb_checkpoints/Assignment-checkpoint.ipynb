```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set(rc={'figure.figsize': [10, 10]}, font_scale=1.2)
```

```python
df=pd.read_csv('50ulke.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.distplot(df['Growth'], kde=False, color='m')
```

```python
sns.distplot(df['Per person'], kde=False, color='m')
```

```python
sns.distplot(df['National Income'], kde=False, color='m')
```

```python
sns.kdeplot(df['National Income'], shade=True, color='w')
```

```python
sns.kdeplot(df['Per person'], shade=True, color='k')
```

```python
sns.kdeplot(df['Growth'], shade=True, color='y')
```

```python
sns.jointplot(x='Growth', y='National Income', data=df, color='m', height=8, ratio=6)
```

```python
sns.jointplot(x='Growth', y='Per person', kind='kde', data=df, color='m', height=8, ratio=6)
```

```python
sns.jointplot(x='National Income', y='Per person', data=df, kind='hex', color='m', height=8, ratio=6)
```

```python
sns.pairplot(df)
```

```python
sns.boxplot(x="Continent", y="Growth", data=df, palette='viridis')
```

```python
sns.boxplot(x="Continent", y="National Income", data=df, palette='viridis')
```

```python
sns.violinplot(x="Continent", y="Per person", data=df, palette='viridis')
```

```python
sns.barplot(x="Continent", y="Per person", data=df, palette="viridis")
```

```python
sns.countplot(x="Continent", data=df, palette='viridis')
```

```python
sns.stripplot(x="Continent", y="Per person", data=df, palette="viridis")
```

```python
sns.swarmplot(x="Continent", y="Per person", data=df, palette="viridis")
```

```python
sns.lmplot(x="Per person", y="National Income", data=df, height=7, aspect=1.3)
```
